```python
# ============================================
# SISTEMA INTELIGENTE GANADERO
# CBTIS 128 - Inteligencia Artificial
# Mario Yañez
# Elabora proyectos de programación lógica
# ============================================

# Lista vacía
ganaderos = []

# ============================================
# FUNCIONES
# ============================================

def agregar_ganadero():

    print("\n--- REGISTRO GANADERO ---")

    tipo = input("Tipo de ganado: ")
    cabezas = input("Número de cabezas: ")
    fierro = input("Registro de fierro: ")
    rancho = input("Nombre del rancho: ")
    produccion = input("Producción (carne/leche): ")
    veterinario = input("Veterinario asignado: ")

    ganadero = {
        "Tipo": tipo,
        "Cabezas": cabezas,
        "Fierro": fierro,
        "Rancho": rancho,
        "Produccion": produccion,
        "Veterinario": veterinario
    }

    ganaderos.append(ganadero)

    print("\n✅ Ganadero agregado correctamente")


def mostrar_ganaderos():

    print("\n============== LISTA DE GANADEROS ==============")

    if len(ganaderos) == 0:

        print("No hay registros")

    else:

        print(f"{'Tipo':<15} {'Cabezas':<12} {'Fierro':<15} {'Rancho':<20} {'Producción':<15} {'Veterinario':<20}")

        print("-" * 100)

        for ganadero in ganaderos:

            print(f"{ganadero['Tipo']:<15} "
                  f"{ganadero['Cabezas']:<12} "
                  f"{ganadero['Fierro']:<15} "
                  f"{ganadero['Rancho']:<20} "
                  f"{ganadero['Produccion']:<15} "
                  f"{ganadero['Veterinario']:<20}")


def buscar_ganado():

    print("\n--- BUSCAR TIPO DE GANADO ---")

    buscar = input("Ingrese el tipo de ganado: ")

    encontrado = False

    for ganadero in ganaderos:

        if ganadero["Tipo"] == buscar:

            print("\n🐄 Registro encontrado")
            print("Rancho:", ganadero["Rancho"])
            print("Veterinario:", ganadero["Veterinario"])

            encontrado = True

    if encontrado == False:
        print("No se encontraron registros")


def inteligencia_artificial():

    print("\n--- IA GANADERA ---")

    produccion = input("Ingrese producción (carne/leche): ")

    if produccion == "carne":

        print("🥩 IA: Se recomienda alimentación alta en proteína")

    elif produccion == "leche":

        print("🥛 IA: Se recomienda control veterinario constante")

    else:

        print("🤖 IA: Producción ganadera registrada")


# ============================================
# MENÚ PRINCIPAL
# ============================================

while True:

    print("\n===================================")
    print("    SISTEMA INTELIGENTE GANADERO")
    print("===================================")

    print("1. Agregar ganadero")
    print("2. Mostrar ganaderos")
    print("3. Buscar tipo de ganado")
    print("4. IA ganadera")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":

        agregar_ganadero()

    elif opcion == "2":

        mostrar_ganaderos()

    elif opcion == "3":

        buscar_ganado()

    elif opcion == "4":

        inteligencia_artificial()

    elif opcion == "5":

        print("Programa finalizado")
        break

    else:

        print("Opción incorrecta")
```
